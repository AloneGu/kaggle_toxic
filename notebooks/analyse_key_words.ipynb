```python
test_data = test_df['comment_text'].values
y_out = curr_best_sub.values[:,1:]
print(test_data[0])
print(y_out[:5])
```

```python
for idx,t in enumerate(test_data):
    for k in low_correct_dict:
        if k.lower() in t.lower():
            print(t)
            old_y = y_out[idx].copy()
            new_res = low_correct_dict[k]
            changed_flag = False
            for j in range(6):
                if new_res[j] == 1:
                    if y_out[idx][j] > 0.5: # update
                        changed_flag = True
                        y_out[idx][j] = min(0.9999,y_out[idx][j]*1.2)
            if changed_flag:
                print(old_y)
                print(y_out[idx])
            print('--------------------')
```
