```python
cls = ['toxic','severe_toxic','obscene','threat','insult','identity_hate']
print(df1.columns)
test_pred = df1[cls].values
print(test_pred[:5])
```

```python
cls_2 = [x+'_oof' for x in cls]
train_pred = df2[cls_2].values
print(train_pred[:5])

with open('../features/wordbatch_feat.pkl','wb') as fout:
    pickle.dump([train_pred,test_pred],fout)
print('done')
```
